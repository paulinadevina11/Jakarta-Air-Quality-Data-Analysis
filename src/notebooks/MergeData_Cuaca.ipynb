# **Merge Data Cuaca Harian dari Berbagai wilayah menjadi 1 file csv**

## Import Library

```python
import pandas as pd
```

## Files mapping for my own comfort :D 

```python
files_map = {
    '../../data/cuaca-harian/cuaca-harian-dki1-bundaranhi.csv': 'Bundaran HI',
    '../../data/cuaca-harian/cuaca-harian-dki2-kelapagading.csv': 'Kelapa Gading',
    '../../data/cuaca-harian/cuaca-harian-dki3-jagakarsa.csv': 'Jagakarsa',
    '../../data/cuaca-harian/cuaca-harian-dki4-lubangbuaya.csv': 'Lubang Buaya',
    '../../data/cuaca-harian/cuaca-harian-dki5-kebonjeruk.csv': 'Kebon Jeruk'
}
```

## Read Data + Add new column "Wilayah"

```python
list_data = []

for file_name, nama_wilayah in files_map.items():
    try:
        df = pd.read_csv(file_name)
        df.insert(0, 'wilayah', nama_wilayah)
        list_data.append(df)

    except FileNotFoundError:
        print(f"File tidak ditemukan: {file_name}")
```

## Merge Dataset

```python
if list_data:
    df_merged = pd.concat(list_data, ignore_index=True)
    df_merged.to_csv('../../data/cuaca-harian/data_cuaca_jakarta.csv', index=False)

print("\nPreviewing:")
df_merged.head()
```
